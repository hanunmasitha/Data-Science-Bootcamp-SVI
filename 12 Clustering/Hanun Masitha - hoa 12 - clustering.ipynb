(1) Lakukan transformasi yang diperlukan pada dataset employee

```python
import pandas as pd
df_employee = pd.read_csv('employee.csv')
df_employee
```

```python
df_employee = df_employee.drop(['Unnamed: 0'], axis=1)
```

```python
missing = []
persen = []

for i in df_employee.columns:
    missing.append(df_employee[i].isnull().sum())
    persen.append(df_employee[i].isnull().sum()/len(df_employee[i])*100)

proporsi_hilang = pd.DataFrame({
        'kolom' : df_employee.columns,
        'missing' : missing,
        'persen_missing' : persen
})
    
proporsi_hilang
```

```python
X_obj = df_employee.select_dtypes('object')
X_num = df_employee.select_dtypes('number')
X_obj
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

oe = OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value=np.nan)
oe.fit(X_obj)
X_train_obj_transform = oe.transform(X_obj)

X_train_obj_transform = pd.DataFrame(X_train_obj_transform)
X_train_obj_transform.columns = X_obj.columns
X_train_obj_transform
```

```python
from sklearn.preprocessing import RobustScaler

minmax_scaler = RobustScaler()
num_minmax_scaled = minmax_scaler.fit_transform(X_num)

num_minmax_scaled = pd.DataFrame(num_minmax_scaled)
num_minmax_scaled.columns = X_num.columns
num_minmax_scaled
```

```python
df_enc = pd.concat([X_train_obj_transform,num_minmax_scaled], axis=1)
```

(2) Lakukan transformasi PCA pada dataset employee. Tentukan jumlah PC yang optimal dengan membuat plot: number of pc vs explained variance ratio. Buat dataframe yang menunjukkan fitur mana yang dominan pada suatu fitur, apabila koefision >0.3, <-0.3 beri warna pada dataframe.

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df_enc)
```

```python
def color(val):
    if val>0.3:
        return 'background: skyblue'
    if val<-0.3:
        return 'background: pink'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x=np.arange(1,11,1), y = pca.explained_variance_ratio_[:10].cumsum(), marker='o')
plt.xlabel('n_components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
pca = PCA(n_components=4)
pca.fit(df_enc)
```

```python
a = {'PC1':pca.components_[0],
             'PC2':pca.components_[1],
             'PC3':pca.components_[2],
             'PC4':pca.components_[3]}
```

```python
b = pd.DataFrame(a)
b.index = df_enc.columns
```

```python
b.style.applymap(color)
```

```python
df_pca = pd.DataFrame(pca.transform(df_enc), columns=['PC1', 'PC2', 'PC3', 'PC4'])
df_pca
```

(3) Dengan menggunakan elbow method dan sillhouette score tentukan nilai k optimal untuk pembuatan model kmeans/kmedoids clustering

```python
from sklearn.cluster import KMeans

x = []
y = []
for k in np.arange(2,11,1):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df_pca)
    x.append(k)
    y.append(kmeans.inertia_)

sns.lineplot(x=x, y=y, marker='o')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

x = []
y = []
for k in np.arange(2,11,1):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df_pca)
    x.append(k)
    y.append(silhouette_score(df_pca, kmeans.labels_))

sns.lineplot(x=x, y=y, marker='o')
plt.show()
```

(4) Buat EDA untuk tiap cluster pada data pra-transformasi (sebelum scaling/encoding): .describe()

```python
from sklearn_extra.cluster import KMedoids
kmeans = KMedoids(n_clusters=3)
kmeans.fit(df_enc)
```

```python
kmeans.cluster_centers_
```

```python
centroid = kmeans.cluster_centers_
```

```python
df_employee2 = df_employee.copy()
df_employee2['label'] = kmeans.labels_
df_employee2
```

```python
for x in range(kmeans.n_clusters):
    print('cluster {}'.format(x))
    print(df_employee2[df_employee2['label']==x].describe())
    print(pd.DataFrame(centroid[x].reshape(1,-1), columns=df_employee2.columns[0:-1]))
    print(list(df_employee2[df_employee2['EmployeeNumber']==x]['Attrition'].unique()))
    print('*'*50)
```

(5) Buat EDA numerik-numerik berupa scatterplot/jointplot dengan label sebagai hue.

```python
df_num = df_employee2.select_dtypes('number')
```

```python
a = []
for x in df_num.columns:
    for y in df_num.columns:
        if x != y:
            if [y,x] not in a:
                print(str(x))
                print(str(y))
                sns.jointplot(x = df_num[x], y = df_num[y])
                plt.show()
                a.append([x,y])
                print('*'*50)
```
