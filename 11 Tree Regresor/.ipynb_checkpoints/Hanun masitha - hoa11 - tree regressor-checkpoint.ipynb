(1) Split X, y dari dataset churn. y = average utilization ratio

```python
import pandas as pd
df_churn = pd.read_csv('churn.csv')
df_churn
```

```python
missing = []
persen = []

for i in df_churn.columns:
    missing.append(df_churn[i].isnull().sum())
    persen.append(df_churn[i].isnull().sum()/len(df_churn[i])*100)

proporsi_hilang = pd.DataFrame({
        'kolom' : df_churn.columns,
        'missing' : missing,
        'persen_missing' : persen
})
    
proporsi_hilang
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.impute import SimpleImputer 

impute = SimpleImputer(strategy = 'most_frequent')
imputer_result = impute.fit_transform(df_churn)

imputer_result = pd.DataFrame(imputer_result)
imputer_result.columns = df_churn.columns

print("imputation on Education_Level")

sns.histplot(df_churn['Education_Level'], alpha=0.2, kde=True)
sns.histplot(imputer_result['Education_Level'], alpha=0.2, kde=True, color='red')
plt.legend(['pre', 'post'])
plt.show()

print("imputation on Marital_Status")

sns.histplot(df_churn['Marital_Status'], alpha=0.2, kde=True)
sns.histplot(imputer_result['Marital_Status'], alpha=0.2, kde=True, color='red')
plt.legend(['pre', 'post'])
plt.show()
```

```python
df_churn['Education_Level'] = imputer_result['Education_Level']
df_churn['Marital_Status'] = imputer_result['Marital_Status']
```

```python
y = df_churn['Avg_Utilization_Ratio']
x = df_churn.drop(['Avg_Utilization_Ratio'], axis = 1)
```

(2) Split trainset dan testset. test size = 0.2.

```python
# Data Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2
)

print("Jumlah Data Train : {0}".format(len(X_train)))
print("Jumlah Data Test : {0}".format(len(X_test)))
```

(3) Lakukan transformasi yang diperlukan pada trainset.

TRAIN DATA

```python
X_train_obj = X_train.select_dtypes('object')
X_train_num = X_train.select_dtypes('number')
X_train_obj
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

oe = OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value=np.nan)
oe.fit(X_train_obj)
X_train_obj_transform = oe.transform(X_train_obj)

X_train_obj_transform = pd.DataFrame(X_train_obj_transform)
X_train_obj_transform.columns = X_train_obj.columns
X_train_obj_transform
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
num_minmax_scaled = minmax_scaler.fit_transform(X_train_num)

num_minmax_scaled = pd.DataFrame(num_minmax_scaled)
num_minmax_scaled.columns = X_train_num.columns
num_minmax_scaled
```

```python
X_train_enc = pd.concat([X_train_obj_transform,num_minmax_scaled], axis=1)
```

DATA TEST

```python
X_test_obj = X_test.select_dtypes('object')
X_test_num = X_test.select_dtypes('number')
X_test_obj
```

```python
X_test_obj_transform = oe.transform(X_test_obj)

X_test_obj_transform = pd.DataFrame(X_test_obj_transform)
X_test_obj_transform.columns = X_test_obj.columns
X_test_obj_transform
```

```python
num_minmax_scaled = minmax_scaler.transform(X_test_num)

num_minmax_scaled = pd.DataFrame(num_minmax_scaled)
num_minmax_scaled.columns = X_test_num.columns
num_minmax_scaled
```

```python
X_test_enc = pd.concat([X_test_obj_transform,num_minmax_scaled], axis=1)
```

(4) Siapkan 3 regressor learning alg.: decision tree regressor, random forest regressor, dan xgboost regressor. Dengan menggunakan cross validation tentukan berbagai hyperparameter yang optimal.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

param_grid = {'max_depth':[3,5,7,9],
             'min_samples_split':[10,20,30,40],
             'max_leaf_nodes':[10,20,30],
             'max_features':[3,5,7]}

gscv = GridSearchCV(DecisionTreeRegressor(), param_grid, cv=5)

gscv.fit(X_train_enc, y_train)
best_param_dtr = gscv.best_params_
```

```python
best_param_dtr
```

```python
tree_def = DecisionTreeRegressor()
tree_opt = DecisionTreeRegressor(**best_param_dtr)
tree_def.fit(X_train_enc, y_train)
tree_opt.fit(X_train_enc, y_train)

y_pred_train_dtr = tree_def.predict(X_train_enc)
y_pred_test_dtr = tree_def.predict(X_test_enc)
y_pred_train_opt_dtr = tree_opt.predict(X_train_enc)
y_pred_test_opt_dtr = tree_opt.predict(X_test_enc)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
performance = {'train-def':{}, 'test-def':{}, 'train-opt':{}, 'test-opt':{}}

for x,y,z in zip(
    [y_pred_train_dtr, y_pred_test_dtr, y_pred_train_opt_dtr, y_pred_test_opt_dtr],
    [y_train, y_test,y_train, y_test],
    ['train-def', 'test-def', 'train-opt', 'test-opt']
):
    performance[z]['r2']=r2_score(y, x)
    performance[z]['mse']=mean_squared_error(y, x)
    performance[z]['rmse']=mean_squared_error(y, x)**(1/2)
    performance[z]['mae']=mean_absolute_error(y, x)
    performance[z]['mape']=mean_absolute_percentage_error(y, x)
    
performance = pd.DataFrame(performance).T
performance.style.highlight_max(color='lightblue', axis =0)
```

```python
from sklearn.ensemble import RandomForestRegressor

param_grid = {'n_estimators':[60,180,300],
              'max_samples':[60,180,300],
              'max_depth':[5,7,9],
             'min_samples_split':[0.1, 0.5, 0.01],
             'max_leaf_nodes':[10,20,30]}


gscv = GridSearchCV(RandomForestRegressor(), param_grid, cv=5, verbose=1)

gscv.fit(X_train_enc, y_train)
best_param_rfr = gscv.best_params_
```

```python
best_param_rfr
```

```python
tree_def_rfr = RandomForestRegressor()
tree_opt_rfr = RandomForestRegressor(**best_param_dtr)
tree_def_rfr.fit(X_train_enc, y_train)
tree_opt_rfr.fit(X_train_enc, y_train)

y_pred_train_rfr = tree_def_rfr.predict(X_train_enc)
y_pred_test_rfr = tree_def_rfr.predict(X_test_enc)
y_pred_train_opt_rfr = tree_opt_rfr.predict(X_train_enc)
y_pred_test_opt_rfr = tree_opt_rfr.predict(X_test_enc)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
performance = {'train-def':{}, 'test-def':{}, 'train-opt':{}, 'test-opt':{}}

for x,y,z in zip(
    [y_pred_train_rfr, y_pred_test_rfr, y_pred_train_opt_rfr, y_pred_test_opt_rfr],
    [y_train, y_test,y_train, y_test],
    ['train-def', 'test-def', 'train-opt', 'test-opt']
):
    performance[z]['r2']=r2_score(y, x)
    performance[z]['mse']=mean_squared_error(y, x)
    performance[z]['rmse']=mean_squared_error(y, x)**(1/2)
    performance[z]['mae']=mean_absolute_error(y, x)
    performance[z]['mape']=mean_absolute_percentage_error(y, x)
    
performance = pd.DataFrame(performance).T
performance.style.highlight_max(color='lightblue', axis =0)
```

```python
from xgboost import XGBRegressor

param_grid = {'n_estimators':[60,180,300],
              'learning_rate': [0.1, 0.5],
              'reg_lambda': [0.1, 1],
              'gamma': [0,5,8],
              'min_child_weight': [0, 1],
              'max_depth':[5,7,9]}


gscv = GridSearchCV(XGBRegressor(), param_grid, cv=5, verbose=1)

gscv.fit(X_train_enc, y_train)
best_param_xgb = gscv.best_params_
```

```python
best_param_xgb
```

```python
tree_def_xgb = XGBRegressor()
tree_opt_xgb = XGBRegressor(**best_param_dtr)
tree_def_xgb.fit(X_train_enc, y_train)
tree_opt_xgb.fit(X_train_enc, y_train)

y_pred_train_xgb = tree_def_xgb.predict(X_train_enc)
y_pred_test_xgb = tree_def_xgb.predict(X_test_enc)
y_pred_train_opt_xgb = tree_opt_xgb.predict(X_train_enc)
y_pred_test_opt_xgb = tree_opt_xgb.predict(X_test_enc)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
performance = {'train-def':{}, 'test-def':{}, 'train-opt':{}, 'test-opt':{}}

for x,y,z in zip(
    [y_pred_train_xgb, y_pred_test_xgb, y_pred_train_opt_xgb, y_pred_test_opt_xgb],
    [y_train, y_test,y_train, y_test],
    ['train-def', 'test-def', 'train-opt', 'test-opt']
):
    performance[z]['r2']=r2_score(y, x)
    performance[z]['mse']=mean_squared_error(y, x)
    performance[z]['rmse']=mean_squared_error(y, x)**(1/2)
    performance[z]['mae']=mean_absolute_error(y, x)
    performance[z]['mape']=mean_absolute_percentage_error(y, x)
    
performance = pd.DataFrame(performance).T
performance.style.highlight_max(color='lightblue', axis =0)
```

(5) Print metrics r2, mse, rmse, mae, dan mape trainset vs testset (dt vs rf vs xgb) dalam bentuk dataframe

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
performance = {'DecisionTreeRegressor-Train':{}, 'DecisionTreeRegressor-Test':{}, 'RandomForestRegressor-Train':{},
               'RandomForestRegressor-Test':{}, 'XGBRegressor-Train':{}, 'XGBRegressor-Test':{}}

for x,y,z in zip(
    [y_pred_train_opt_dtr, y_pred_test_opt_dtr, 
     y_pred_train_opt_rfr, y_pred_test_opt_rfr, 
     y_pred_train_opt_xgb, y_pred_test_opt_xgb],
    [y_train, y_test,y_train, y_test ,y_train, y_test],
    ['DecisionTreeRegressor-Train', 'DecisionTreeRegressor-Test', 'RandomForestRegressor-Train', 'RandomForestRegressor-Test',
     'XGBRegressor-Train', 'XGBRegressor-Test']
):
    performance[z]['r2']=r2_score(y, x)
    performance[z]['mse']=mean_squared_error(y, x)
    performance[z]['rmse']=mean_squared_error(y, x)**(1/2)
    performance[z]['mae']=mean_absolute_error(y, x)
    performance[z]['mape']=mean_absolute_percentage_error(y, x)
    
performance = pd.DataFrame(performance).T
performance.style.highlight_max(color='lightblue', axis =0)
```

Jika dilihat dari 5 perhitungan (R2, MSE, RMSE, MAE, MAPE) dapat disimpulkan bahwa :
    1. Decision Tree Regressor mendapatkan nilai terbaik dengan menggunakan MSE, RMSE, dan MAR
    2. Random Forest Regressor mendapatkan nilai terbaik dengan menggunakan MAPE
    3. XGB Regressor mendapatkan nilai terbaik dengan menggunakan R2
    
dari nilai diatas dapat disimpulkan bahwa Decision Tree Regressor adalah regresor learning algoritma teroptimal dari pada yang lain
