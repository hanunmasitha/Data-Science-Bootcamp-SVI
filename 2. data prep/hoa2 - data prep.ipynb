4. lakukan imputation pada kolom dengan missing value pada df dengan aturan:<p>
    <p>
    a. missing value â‰¤10% gunakan simple imputer seperti median<p>
    b. missing value >10% dan â‰¤30%  gunakan advance imputer seperti KNN atau MICE <p>
    c. missing value >30% drop kolom

```python
missing = []
persen = []

for i in df.columns:
    missing.append(df[i].isnull().sum())
    persen.append(df[i].isnull().sum()/len(df[i])*100)

proporsi_hilang = pd.DataFrame({
        'kolom' : df.columns,
        'missing' : missing,
        'persen_missing' : persen
})
    
proporsi_hilang
```

dari hasil diatas dapat disimpulkan bahwa ada missing value pada
    1. Age dengan 8%
    2. EducationField dengan 15%
    3. MonthlyIncome dengan 21%
    4. TotalWorkingYears dengan 19%

```python
import missingno as msno

msno.matrix(df)
```

```python
df_hasil = df.copy(deep=True)
```

Missing value < 10% adalah Age. sehingga penyelesaiannya dengan menggunakan simple imputer

```python
from sklearn.impute import SimpleImputer 

impute = SimpleImputer(strategy = 'median')
```

```python
df2 = df.copy(deep=True)
df2 = df2.select_dtypes(include='number')
df2
```

```python
impute.fit(df2)
df2_imputed = impute.transform(df2)
```

```python
df2_imputed = pd.DataFrame(df2_imputed)
df2_imputed
```

```python
df_hasil['Age'] = df2_imputed[1]
msno.matrix(df_hasil)
```

Missing value 10% < x < 30% adalah EducationField, MonthlyIncome, TotalWorkingYears. sehingga penyelesaiannya dengan menggunakan advance imputer

```python
df2 = df.copy(deep=True)
df2['EducationField'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

ord_enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value = -1)
#x["job"].dropna().values.reshape(-1,1)

df2.loc[df2.EducationField.notna(), ['EducationField']]= ord_enc.fit_transform(df2['EducationField'].dropna().values.reshape(-1,1))
df2['EducationField']
```

```python
ord_enc.categories_
```

```python
df2['MonthlyIncome'].unique()
```

```python
df2['TotalWorkingYears'].unique()
```

```python
df2 = df2.loc[:, ['EducationField', 'MonthlyIncome', 'TotalWorkingYears']]
df2
```

```python
from sklearn.impute import KNNImputer

imputer_knn = KNNImputer(n_neighbors=5)
imputer_knn.fit(df2)
df_knn = imputer_knn.transform(df2)
```

```python
df_knn = pd.DataFrame(df_knn)
df_knn.columns = df2.columns
df_knn
```

```python
df_hasil['EducationField'] = df_knn['EducationField']
df_hasil['MonthlyIncome'] = df_knn['MonthlyIncome']
df_hasil['TotalWorkingYears'] = df_knn['TotalWorkingYears']

msno.matrix(df_hasil)
```

```python
df_hasil['EducationField']
```

```python
df_hasil['EducationField'] = df_hasil['EducationField'].round()
df_hasil['EducationField']
```

```python
df_hasil['EducationField'] = ord_enc.inverse_transform(df_hasil[['EducationField']].astype(str))
df_hasil['EducationField']
```

5. Untuk tiap kolom yang dilakukan transformasi buatlah histogram pre dan post imputation. Print mean dan std pre dan post imputation.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

print("imputation on age")

sns.histplot(df['Age'], alpha=0.2, kde=True)
sns.histplot(df_hasil['Age'], alpha=0.2, kde=True, color='red')
plt.legend(['pre', 'post'])
plt.show()
print('pre: mean: {}, std: {}'.format(np.mean(df['Age']), 
                                      np.std(df['Age'])))
print('post: mean: {}, std: {}'.format(np.mean(df_hasil['Age']), 
                                      np.std(df_hasil['Age'])))
```
