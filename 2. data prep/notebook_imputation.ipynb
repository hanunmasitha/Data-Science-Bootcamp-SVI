1. lakukan imputation pada kolom 'university rating' dengan menggunakan simple imputer dengan strategi median

bandingkan distribusi data antara pre dan post-imputation

- buat histogram sebelum impute
- buat histrogram setelah impute
- di satu graph

print(mean pre:, std pre:)
print(mean post:, std post:)

```python
admit2 = admit_miss.copy(deep=True)
```

```python
impute2 = SimpleImputer(strategy='median')
```

```python
impute2.fit(admit2)
```

```python
admit2_imputed = impute2.transform(admit2)
```

```python
admit2_imputed = pd.DataFrame(admit2_imputed)
admit2_imputed.columns = admit2.columns
```

```python
import seaborn as sns
```

```python
admit2.head()
```

```python
import matplotlib.pyplot as plt
sns.histplot(admit2['University Rating'], alpha=0.2, kde=True)
sns.histplot(admit2_imputed['University Rating'], alpha=0.2, kde=True, color='red')
plt.legend(['pre', 'post'])
plt.show()
print('pre: mean: {}, std: {}'.format(np.mean(admit2['University Rating']), 
                                      np.std(admit2['University Rating'])))
print('post: mean: {}, std: {}'.format(np.mean(admit2_imputed['University Rating']), 
                                      np.std(admit2_imputed['University Rating'])))
```

```python
df2_knn = pd.DataFrame(df2_knn)
```

```python
df2_knn.columns = df2.columns
```

```python
msno.matrix(df2_knn)
```

```python
msno.matrix(df2)
```

```python
import matplotlib.pyplot as plt
sns.histplot(admit2['University Rating'], alpha=0.2, kde=True)
sns.histplot(admit2_imputed['University Rating'], alpha=0.2, kde=True, color='red')
plt.legend(['pre', 'post'])
plt.show()
print('pre: mean: {}, std: {}'.format(np.mean(admit2['University Rating']), 
                                      np.std(admit2['University Rating'])))
print('pre: mean: {}, std: {}'.format(np.mean(admit2_imputed['University Rating']), 
                                      np.std(admit2_imputed['University Rating'])))
```
